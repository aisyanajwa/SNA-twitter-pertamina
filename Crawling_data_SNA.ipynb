**keywords**: List kata kunci yang digunakan dalam pencarian, berhubungan dengan kasus korupsi Pertamina.

**hashtags**: List hashtag yang relevan dengan topik yang akan diambil dari Twitter.

Menyusun format pencarian untuk tweet-harvest:
1. "query": Kata kunci atau hashtag yang
dicari.
2. since:2025-02-25 until:2025-03-12: Mencari tweet dalam rentang 25 Februari 2025 - 12 Maret 2025.
3. lang:id: Hanya mencari tweet dalam bahasa Indonesia.

Menjalankan perintah scraping menggunakan tweet-harvest melalui os.system().

Parameter:
1. npx -y tweet-harvest@2.6.1: Menjalankan tweet-harvest versi 2.6.1.
2. -o "{filename}": Menyimpan hasil dalam file CSV yang telah ditentukan.
3. -s "{search_query}": Menggunakan kata kunci yang telah dirancang sebelumnya.
4. --tab "LATEST": Mengambil tweet terbaru dari Twitter.
5. -l 1500: Mengambil maksimal 1500 tweet untuk setiap pencarian.
6. --token {twitter_auth_token}: Menggunakan token autentikasi untuk mengakses data Twitter.



```python
import os

keywords = [
    "Korupsi Pertamina", "Kasus Korupsi Pertamina", "Mafia Migas",
    "Direksi Pertamina", "Skandal Keuangan Pertamina", "KPK Pertamina",
    "Audit Pertamina", "BUMN korupsi", "Kasus suap Pertamina",
    "Komisaris Pertamina", "Investigasi KPK Pertamina", "Menteri BUMN korupsi"
]

hashtags = [
    "#KorupsiPertamina", "#MafiaMigas", "#BUMNBersih", "#SaveKPK",
    "#SkandalPertamina", "#AuditPertamina"
]

query_list = keywords + hashtags

for i, query in enumerate(query_list):
    filename = f'korupsi_pertamina_{i+1}.csv'
    search_query = f'"{query}" since:2025-02-25 until:2025-03-12 lang:id'

    print(f"Mulai crawling: {query}")

    os.system(f'npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_query}" --tab "LATEST" -l 1500 --token {twitter_auth_token}')

    print(f"Crawling selesai untuk: {query}, data tersimpan di {filename}")

print("Semua crawling selesai!")
```

## Data Understanding

Ini adalah contoh dataset yang didapat dari proses scrapping di atas: Click [Dataset](https://drive.google.com/drive/folders/1nikEPSsWf-CiPWDVvMAynd6WnV1JLG8O?usp=sharing)



Dataset ini berisi data tweet yang diperoleh dari pencarian dalam rentang waktu 25 Februari 2025 - 12 Maret 2025. Data ini mencakup berbagai informasi terkait tweet, seperti teks tweet, tanggal, jumlah like, reply, retweet, username, lokasi, dan URL tweet.

1. conversation_id_str	ID percakapan tweet yang menunjukkan jika tweet tersebut bagian dari sebuah thread.
2. created_at	Tanggal dan waktu tweet diposting dalam format UTC.
3. favorite_count	Jumlah like (favorit) pada tweet.
4. full_text	Isi lengkap dari tweet.
5. id_str	ID unik dari tweet.
6. image_url	URL gambar yang disertakan dalam tweet (jika ada).
7. in_reply_to_screen_name	Username yang dituju dalam reply (jika merupakan balasan).
8. lang	Bahasa yang digunakan dalam tweet.
9. location	Lokasi pengguna (jika tersedia).
10. quote_count	Jumlah tweet lain yang mengutip tweet ini.
11. reply_count	Jumlah balasan pada tweet.
12. retweet_count	Jumlah retweet yang diterima.
13. tweet_url	URL langsung menuju tweet di Twitter.
14. user_id_str	ID unik pengguna yang membuat tweet.
15. username	Nama pengguna (handle Twitter) dari akun yang memposting tweet.

## Merge

```python
import pandas as pd

total_data = 0

for i in range(1, 15):
    filename = f'/content/drive/MyDrive/MAKUL/SNA/Dataset/korupsi_pertamina_{i}.csv'
    try:
        df = pd.read_csv(filename)
        jumlah_baris = len(df)
        print(f'Jumlah data di {filename}: {jumlah_baris}')
        total_data += jumlah_baris
    except pd.errors.EmptyDataError:
        print(f'File {filename} is empty or unreadable. Skipping...')

print(f'\nTotal jumlah data dari seluruh dataset: {total_data}')
```

```python
import pandas as pd
df_main = pd.concat([pd.read_csv(f'/content/drive/MyDrive/MAKUL/SNA/Dataset/korupsi_pertamina_{i}.csv') for i in range(1, 15)])
```
